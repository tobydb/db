```python
import numpy as np

class Adaline:

    def __init__(self, learning_rate=0.01, n_iterations=100):

        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None

    def fit(self, X, y):

        self.weights = np.zeros(1 + X.shape[1])

        for _ in range(self.n_iterations):

            net_input = np.dot(X, self.weights[1:]) + self.weights[0]

            predicted = net_input

            errors = y - predicted

            self.weights[1:] += self.learning_rate * np.dot(X.T, errors)

            self.weights[0] += self.learning_rate * np.sum(errors)

    def predict(self, X):

        return np.dot(X, self.weights[1:]) + self.weights[0]


X = np.array([[1,2], [2,3], [3,4]])

y = np.array([1,1,0])

adaline = Adaline(learning_rate=0.01,n_iterations=10)

adaline.fit(X, y)

print("Weights:", adaline.weights)

print("Prediction:", adaline.predict(np.array([2,3])))

# output:
# Weights: [0.08127088 0.03660716 0.11787805]
# Prediction: 0.5081193508836386
```
